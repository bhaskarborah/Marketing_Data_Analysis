```python
# import dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# load the data
df = pd.read_csv("Resources/market_complete.csv")
df
```

```python
# List all columns & datatypes
df.dtypes
```

```python
# Drop columns that are not useful to build a logistic regression model
df = df[['Income', 'mnt_sum', 'purchases_sum', 'Kidhome','NumWebVisitsMonth', 'Total_campaigns', 'Response']]
df
```

```python
df.columns
```

```python
# Transform text into numerical data
le = LabelEncoder()
df['Education'] = le.fit_transform(df['Education'])
df['Marital_Status'] = le.fit_transform(df['Marital_Status'])
df['Country'] = le.fit_transform(df['Country'])
df
```

```python
# Separate the features from the target
y = df["Response"]
X = df.drop(columns="Response")
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
X_train.shape
```

```python
# Create a logistic regression model
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=200,
                                random_state=1)
```

```python
# Fit (train) the model
classifier.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = classifier.predict(X_test)
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
results.head(20)
```

```python
# Evaluate predictions
print(accuracy_score(y_test, y_pred))
```
